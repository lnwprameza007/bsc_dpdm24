```python
# prompt: find most invoice by country

import pandas as pd

# Assuming 'data' DataFrame is already loaded as shown in your provided code

invoice_counts_by_country = data.groupby('Country')['InvoiceNo'].count()
most_invoices_country = invoice_counts_by_country.idxmax()
print(f"The country with the most invoices is: {most_invoices_country}")
```

```python
# prompt: แบ่งเวลาเช้า กลางวันเย็น สร้างคอลัมภ์ใหม่

# Assuming 'data' DataFrame is already loaded as shown in your provided code
import pandas as pd

# Convert 'InvoiceDate' to datetime objects if it's not already
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])

# Extract the hour from the 'InvoiceDate'
data['Hour'] = data['InvoiceDate'].dt.hour

# Create the 'TimeOfDay' column
def categorize_time_of_day(hour):
    if 6 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 18:
        return 'Afternoon'
    else:
        return 'Evening'

data['TimeOfDay'] = data['Hour'].apply(categorize_time_of_day)

data
```

```python
# prompt: select country = United Kingdom'

country = 'United Kingdom' #@param ["United Kingdom", "France", "Australia", "Netherlands", "Germany", "Norway", "EIRE", "Switzerland", "Spain", "Poland", "Portugal", "Italy", "Belgium", "Lithuania", "Japan", "Iceland", "Channel Islands", "Denmark", "Cyprus", "Sweden", "Finland", "Austria", "Greece", "Hong Kong", "Singapore", "Lebanon", "United Arab Emirates", "Israel", "Saudi Arabia", "Czech Republic", "Canada", "Brazil", "USA", "European Community", "Bahrain", "Malta", "RSA", "Unspecified"]

# Now you can use the selected country in your code
print(f"Selected country: {country}")

# Example usage with the selected country
uk_data = data[data['Country'] == country]
print(uk_data.head())
```

```python
morning = uk_data[uk_data['TimeOfDay'] == 'Morning']
afternoon = uk_data[uk_data['TimeOfDay'] == 'Afternoon']
evening = uk_data[uk_data['TimeOfDay'] == 'Evening']
```

```python
tran_morning = morning.groupby('CustomerID')['StockCode'].apply(list)
tran_afternoon = afternoon.groupby('CustomerID')['StockCode'].apply(list)
tran_evening = evening.groupby('CustomerID')['StockCode'].apply(list)
```

```python
min_sup = 0.055

frequentItemsets = list(apriori(tran_morning,min_support = min_sup))
for i in frequentItemsets:
    if len(i[0])>1:
        print('yes')
        print(i)
```

```python
min_sup = 0.07

frequentItemsets = list(apriori(tran_afternoon,min_support = min_sup))
for i in frequentItemsets:
    if len(i[0])>1:
        print('yes')
        print(i)
```

```python
min_sup = 0.04

frequentItemsets = list(apriori(tran_evening,min_support = min_sup))
for i in frequentItemsets:
    if len(i[0])>1:
        print('yes')
        print(i)
```
