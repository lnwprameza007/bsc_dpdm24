##JSON

```python
# prompt: read json file name data_station in drive

import json
import pandas as pd

def read_json_from_drive(file_path):
    try:
        with open(file_path, 'r') as f:
            data = json.load(f)
        return pd.DataFrame(data)  # Convert JSON data to DataFrame
    except FileNotFoundError:
        print(f"Error: File not found at {file_path}")
        return None
    except json.JSONDecodeError:
        print(f"Error: Invalid JSON format in file {file_path}")
        return None
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return None

# Example usage (replace with your actual file path)
json_file_path = "/content/drive/My Drive/bsc_dpdm24/data/data_station.json"
df_json = read_json_from_drive(json_file_path)

if df_json is not None:
    print(df_json.head())
```
