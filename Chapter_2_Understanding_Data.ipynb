##JSON

```python
import json
import pandas as pd

def read_json_from_drive(file_path):
    try:
        with open(file_path, 'r') as f:
            data = json.load(f)
        return pd.DataFrame(data)  # Convert JSON data to DataFrame
    except FileNotFoundError:
        print(f"Error: File not found at {file_path}")
        return None
    except json.JSONDecodeError:
        print(f"Error: Invalid JSON format in file {file_path}")
        return None
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return None

# Example usage (replace with your actual file path)
json_file_path = "/content/drive/My Drive/bsc_dpdm24/data/data_station.json"
df_json = read_json_from_drive(json_file_path)

if df_json is not None:
    print(df_json.head())
```

#Distance

```python
import pandas as pd

data = [[1, 2], [-2, 4], [0, 3], [-3, -2]]
df = pd.DataFrame(data, columns=['X1', 'X2'])
df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6)) # Fixed: Removed extra indentation
plt.scatter(df['X1'], df['X2']) # Fixed: 'X1' and 'X2' should be strings
plt.plot([-3,1],[0,0],color='black')
plt.plot([0,0],[-2,4],color='black')
plt.xlabel("X1")
plt.ylabel("X2")
plt.title("Scatter Plot of X1 vs X2")
plt.grid(True)
plt.show()
```

```python
from scipy.spatial.distance import pdist, squareform

distance_matrix = squareform(pdist(df))

distance_matrix
```

```python
from scipy.spatial.distance import pdist, squareform

distance_matrix = squareform(pdist(df, 'cityblock')) # l1 norm is cityblock

distance_matrix
```

#Normalization (Z-score & Min-max normalization)

##Min-max

```python
df2_normalized = df2.copy()

for column in df2_normalized.columns:
    # Calculate min and max for the column
    min_val = df2_normalized[column].min()
    max_val = df2_normalized[column].max()

    # Apply min-max normalization
    df2_normalized[column] = (df2_normalized[column] - min_val) / (max_val - min_val)

df2_normalized
```

##Z-score

```python
from scipy.stats import zscore
df2_zscore = df2.apply(zscore)
df2_zscore
```
