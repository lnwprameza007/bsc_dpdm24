##แปลง data เนื่องจากโมเดล ไม่สามารถใช้ string ได้

```python
apndx_data.info()
```

```python
from sklearn.preprocessing import LabelEncoder

# List of columns to transform
columns_to_transform = [
    'Smoking_Status', 'Alcohol_Consumption', 'Family_History_Cancer',
    'Genetic_Mutations', 'Physical_Activity_Level', 'Radiation_Exposure',
    'Previous_Cancers', 'Tumor_Markers','Symptom_Severity'
]

# Apply label encoding to each specified column
encoder = LabelEncoder()
for column in columns_to_transform:
    apndx_data[column] = encoder.fit_transform(apndx_data[column].astype(str))

# Show the transformed columns for verification
apndx_data.info()
```

```python
apndx_data.head()
```

```python
# Use one-hot encoding for 'Gender', 'Chronic_Diseases', and 'Diet_Type'
from sklearn.preprocessing import OneHotEncoder

# Create a OneHotEncoder object
enc = OneHotEncoder(handle_unknown='ignore')

# Fit and transform the specified columns
encoded_features = enc.fit_transform(apndx_data[['Gender', 'Chronic_Diseases', 'Diet_Type']]).toarray()

# Create new column names for the encoded features
feature_names = enc.get_feature_names_out(['Gender', 'Chronic_Diseases', 'Diet_Type'])

# Create a new DataFrame with the encoded features
encoded_df = pd.DataFrame(encoded_features, columns=feature_names)

# Concatenate the encoded DataFrame with the original DataFrame
apndx_data = pd.concat([apndx_data, encoded_df], axis=1)

# Drop the original columns
apndx_data = apndx_data.drop(['Gender', 'Chronic_Diseases', 'Diet_Type'], axis=1)

# Display the updated DataFrame
apndx_data.head()
```

```python
mapping = {'No': 0, 'Yes': 1}
apndx_data['Appendix_Cancer_Prediction'] = apndx_data['Appendix_Cancer_Prediction'].map(mapping)
apndx_data['Appendix_Cancer_Prediction'].unique()
```

#เก็บ Test เอาไว้วัดตอนจบ

```python
from sklearn.model_selection import train_test_split

# Assuming apndx_data is your DataFrame
train_data, test_data = train_test_split(apndx_data, train_size=200000, random_state=9)

print("Training data shape:", train_data.shape)
print("Testing data shape:", test_data.shape)
```

#เลือก Model ที่ดีที่สุด(Model Selection)

1.Decision Trees

2.(Short)Decetion Trees

3.Naive Bayes

####split data into 10 folds

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=10, shuffle=True, random_state=6)
```

####Defind X,y

```python
X = train_data.drop(columns=['Appendix_Cancer_Prediction'])
y = train_data['Appendix_Cancer_Prediction']
```

##Decision Trees

```python
#import
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score

#defind
dt_classifier = DecisionTreeClassifier() #default

#train #evaluation
cv_scores = cross_val_score(dt_classifier, X, y, cv=kf, scoring='accuracy')
```

```python
#Print Cross-validation scores
print("Cross-validation scores:", cv_scores)
print("Average cross-validation score:", cv_scores.mean())
print("Standard deviation of cross-validation scores:", cv_scores.std())
```

##(Short)Decision Trees

```python
#import
from sklearn.tree import DecisionTreeClassifier

#defind
dt_classifier_short = DecisionTreeClassifier(max_depth=5)

#train #evaluation
cv_scores_short = cross_val_score(dt_classifier_short, X, y, cv=kf, scoring='accuracy')

#Print Cross-validation scores
print("Cross-validation scores:", cv_scores_short)
print("Average cross-validation score:", cv_scores_short.mean())
print("Standard deviation of cross-validation scores:", cv_scores_short.std())
```

##Naive Bayes

```python
#import
from sklearn.naive_bayes import GaussianNB

#defind
nb_classifier = GaussianNB()

#train #evaluation
cv_scores_nb = cross_val_score(nb_classifier, X, y, cv=kf, scoring='accuracy')

# Print Cross-validation scores
print("Cross-validation scores (Naive Bayes):", cv_scores_nb)
print("Average cross-validation score (Naive Bayes):", cv_scores_nb.mean())
print("Standard deviation of cross-validation scores (Naive Bayes):", cv_scores_nb.std())
```

#เลือก Model ที่แม่นที่สุด คือ Naive Bayes มา train ด้วย data ทั้งหมด

```python
#import
from sklearn.naive_bayes import GaussianNB

#defind
nb_classifier = GaussianNB()

#train
nb_classifier.fit(X, y)
```

##วัดผลด้วย test data

```python
X_test = test_data.drop(columns=['Appendix_Cancer_Prediction'])
y_test = test_data['Appendix_Cancer_Prediction']
```

```python
#Predict
y_pred = nb_classifier.predict(X_test)
```

```python
#Evaluate
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy on test data: {accuracy}")
```
